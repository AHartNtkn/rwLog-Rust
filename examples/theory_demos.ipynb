# CHR Theory Demos

Load the example theory files and run sample queries.


```rwlog
load equality_constraints.txt

rel same {
    (pair $x $y) { (eq $x $y) } -> $x
}

rel different {
    (pair $x $y) { (neq $x $y) } -> (pair $x $y)
}

rel nonzero_ok {
    $x { (nonzero $x) } -> $x
}
```

```rwlog
# Equal numbers
@(pair (s (s z)) (s (s z))) ; same
next
```

```rwlog
# Different numbers
@(pair (s z) z) ; different
next
```

```rwlog
# Guarded nonzero check
@(s z) ; nonzero_ok
next
```

```rwlog
load range_sets.txt

rel pred {
    (s $x) -> $x
}

rel close_range {
    (range $lo (open z)) -> empty
    |
    (range (closed $lo) (closed $hi)) -> (range (closed $lo) (closed $hi))
    |
    (range (open $lo) (closed $hi)) -> (range (closed (s $lo)) (closed $hi))
    |
    [(range (closed $lo) (open $hi)) -> $hi ; pred ; $hi2 -> (range (closed $lo) (closed $hi2))]
    |
    [(range (open $lo) (open $hi)) -> $hi ; pred ; $hi2 -> (range (closed (s $lo)) (closed $hi2))]
}

rel card_range {
    empty -> z
    |
    (range (closed $x) (closed $x)) -> (s z)
    |
    (range (closed $lo) (closed $hi)) { (lt $hi $lo) } -> z
    |
    (range (closed $lo) (closed $hi)) { (lt $lo $hi) }
        -> (range (closed (s $lo)) (closed $hi)) ; card_range ; $n -> (s $n)
}

rel card {
    close_range ; card_range
}

rel leq_ok {
    (pair $x $y) { (leq $x $y) } -> (pair $x $y)
}

rel member {
    (pair $x (rcons $range $rest)) { (between $x $range) } -> $x
    |
    (pair $x (rcons $range $rest)) -> (pair $x $rest) ; member
}
```

```rwlog
# Membership in a range set
@(pair (s (s z)) (rcons (range (closed z) (open (s (s (s z))))) rnil)) ; member
next
```

```rwlog
# Cardinality of [0,3)
@(range (closed z) (open (s (s (s z))))) ; card
next
```

```rwlog
# Ordering on Peano numbers
@(pair (s z) (s (s z))) ; leq_ok
next
```
