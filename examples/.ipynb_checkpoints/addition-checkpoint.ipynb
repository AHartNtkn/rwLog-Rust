# rwlog addition demo

This notebook loads the Peano addition relation and runs forward/backward queries.


```rwlog
load examples/addition.txt
```

```rwlog
# 0 + 1 = 1
?- @(cons z (s z)) ; add
```

```rwlog
# 3 + 2 = 5
?- @(cons (s (s (s z))) (s (s z))) ; add
```

```rwlog
# Backwards: all pairs that sum to 5
?- add ; @(s (s (s (s (s z)))))
```
