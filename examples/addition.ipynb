# rwlog addition demo

This definition is recursive and can run forward or backward.
This notebook loads the Peano addition relation and runs forward/backward queries.
If loading fails, adjust the path to match your working directory.


```rwlog
load addition.txt
```

```rwlog
# 0 + 1 = 1
@(cons z (s z)) ; add
next
```

```rwlog
# 3 + 2 = 5
@(cons (s (s (s z))) (s (s z))) ; add
next
```

```rwlog
# Backwards: all pairs that sum to 5
add ; @(s (s (s (s (s z)))))
next
next
next
next
next
next
```

```rwlog
# Next answer
add
next
next
next
```
